```python
"""

Creating new df with relevant columns for the map.
Add constant column 'deaths', to summarize later.

"""
map_df = clean_df[['state', 'state_ID', 'race']]
map_df['deaths'] = 1
map_df
map_df['race'].to_string()
```

```python
#map_df['race'].value_counts()
map_df11 = map_df[['state', 'race']]
```

```python
all_categories = [0,1,2,3,4,5] 
map_df11 = pd.get_dummies(map_df11.race).reindex(columns=all_categories, fill_value=0)
```

```python
map_df = map_df.join(map_df11)
```

```python
map_df2=map_df[['state', 0,1,2,3,4,5]]
map_df3 = map_df2.groupby('state').sum()
map_df4 = map_df3.idxmax(axis=1)
```

```python
"""lave navne til race number som kan joines"""
map_df3 = map_df3.reset_index(drop=False)
map_df3
```

```python
ddff = map_df3.set_axis(['State','Asian', 'Black', 'Hispanic', 'Native', 'Others', 'White'], axis=1, inplace=False)
ddff = ddff.reset_index(drop=True)
ddff
```

```python
ddff
```

```python
df = pd.DataFrame(data=map_df4)
df.columns =[ 'max_race']
df = df.reset_index(drop=False)
df["max_race"] = df["max_race"].astype("category")

```

```python
df = df.join(ddff)
df
```

```python
 """
A = 0 
B = 1
H = 2
N = 3
0 = 4
W = 5
"""
df['race_name'] = '' 

def addRace(df):

    for index in df.index:
        if df.loc[index, 'max_race'] == 0:
            #df.loc[df.index[index], 'flee_boolean'] = False
            df.loc[df.index[index], 'race_name'] = 'Asian'
            
        elif df.loc[index, 'max_race'] == 1:
            #df.loc[df.index[index], 'flee_boolean'] = True
            df.loc[df.index[index], 'race_name'] = 'Black'
            
        elif df.loc[index, 'max_race'] == 2:
            #df.loc[df.index[index], 'flee_boolean'] = True
            df.loc[df.index[index], 'race_name'] = 'Hispanic'
            
        elif df.loc[index, 'max_race'] == 3:
            #df.loc[df.index[index], 'flee_boolean'] = True
            df.loc[df.index[index], 'race_name'] = 'Native'
            
        elif df.loc[index, 'max_race'] == 4:
            #df.loc[df.index[index], 'flee_boolean'] = True
            df.loc[df.index[index], 'race_name'] = 'Other'
        
        elif df.loc[index, 'max_race'] == 5:
            #df.loc[df.index[index], 'flee_boolean'] = True
            df.loc[df.index[index], 'race_name'] = 'White'

```

```python
"""

Create the data object for plotting need to be a list of dictionaries.

"""

data = [ dict(
            type='choropleth', # type of map-plot
            colorscale = scl,
            autocolorscale = False,
            locations = df['state'], # the column with the state
            z = df['max_race'].astype(str), # the variable I want to color-code
            locationmode = 'USA-states',
            text =  "<sup>Top race victim of fatal US police shootings in state</sup> <br>Race: "+ df['race_name'].apply(str),
            marker = dict(     # for the lines separating states
                        line = dict (
                                  color = 'rgb(255,255,255)', 
                                  width = 2) ),               
            colorbar = dict(
                        title = "Fatal US Police Shootings <br><sup>From January 1, 2015 to September 7, 2020</sup>")
            ) 
       ]
#  text =  "<sup>total fatal police shootings in state</sup> <br>State ID: "+ map_df['state_ID'].apply(str), #map_df['state_ID'], # hover text
```

```python
"""

Then take care of the layout and plot the map.

"""


layout = dict(
        geo = dict(
            scope='usa',
            projection=dict( type='albers usa' ),        
               ),
             )
fig = dict( data=data, layout=layout )
plotly.offline.iplot(fig)

```

```python
"""

Opens the map in browser.

"""

plotly.offline.plot(fig, filename='fatalPoliceShootings.html')

#py.plot(fig, filename='fatalPoliceShootings.html') 
```
